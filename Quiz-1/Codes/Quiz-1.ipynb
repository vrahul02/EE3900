```python
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt


x_0=[]
n=np.linspace(-2,3,6)

x_0.append(0)
x_0.append(1)
x_0.append(2)
x_0.append(1)
x_0.append(0)
x_0.append(0)

plt.stem(n,x_0,use_line_collection=True)
plt.title('$x_0[n]$')
plt.xlabel('$n$')
plt.ylabel('$x_0(n)$')
plt.grid()
```
